```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

### Functional Method

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
# Multi-Plot on same canvas
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Object-oriented method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(x,y,'g')
axes2.set_title('Smaller Plot')
```

```python
### Different method
```

```python
fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3)
for current_ax in axes:
    current_ax.plot(x,y)
    
```

```python
fig, rows = plt.subplots(nrows=2, ncols=2)
i = 0
for row in rows:
    for ax in row:
        i += 1
        ax.plot(x, y, 'g')
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title('Axes No.{}'.format(i))
fig    
plt.tight_layout()
```

```python
fig.savefig('my_test.png',dpi=200)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

### Figure Size and DPI

```python
fig = plt.figure(figsize=(5,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,'r') 
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(5,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=0)
```
